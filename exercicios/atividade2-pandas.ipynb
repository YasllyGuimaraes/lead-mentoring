07) Adicione uma coluna com o resultado da relação da diferença entre o censo de 2010 e a população estimada em 2017?
- Formula = (2017-2010)/2010

```python
population_df['2017 - 2010'] = population_df['2017PopEstimate'] - population_df['2010Census'] / population_df['2010Census']
population_df['2017 - 2010']

ax = sns.factorplot(x='2017 - 2010', y='County', data=population_df)
ax.figure.set_size_inches(15, 10)
```
