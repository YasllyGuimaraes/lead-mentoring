# 5

Crie 3 vetores de atributos distintos, e utilizando as distâncias euclidiana e Manhattan, determine a qual classe cada vetor de atributos pertence.

```python
from scipy.spatial import distance
```

```python
df['classes'] = list(iris.target)
df
```

```python
classe_0 = df[df['classes'] == 0]
classe_0.drop('classes', axis=1, inplace=True)
vetor_0 = list(classe_0.mean())

print(vetor_0)
```

```python
classe_1 = df[df['classes'] == 1]
classe_1.drop('classes', axis=1, inplace=True)
vetor_1 = list(classe_1.mean())

print(vetor_1)

```

```python
classe_2 = df[df['classes'] == 2]
classe_2.drop('classes', axis=1, inplace=True)
vetor_2 = list(classe_2.mean())

print(vetor_2)
```

Para setosa:

```python
my_vector = [3, 5, 1, 2]
de = distance.euclidean(my_vector, vetor_0)
print("Distancia euclidiana:", de)

dm = distance.cityblock(my_vector, vetor_0)
print("Distancia de manhatan:", dm)
```

Para versicolor:

```python
my_vector = [3, 5, 1, 2]
de = distance.euclidean(my_vector, vetor_1)
print("Distancia euclidiana:", de)

dm = distance.cityblock(my_vector, vetor_1)
print("Distancia de manhatan:", dm)
```

Para virginica:

```python
my_vector = [3, 5, 1, 2]
de = distance.euclidean(my_vector, vetor_2)
print("Distancia euclidiana:", de)

dm = distance.cityblock(my_vector, vetor_2)
print("Distancia de manhatan:", dm)
```

A classe que mais se aproxima do vetor escolhido é a classe setosa.
