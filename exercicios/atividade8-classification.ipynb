Realize a classificação utilizando pelo menos os seguintes métodos:
- kNN
- Decision Trees

- MLP (uma camada oculta)
- MLP (duas camadas ocultas)

```python
from sklearn.naive_bayes import GaussianNB, CategoricalNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, f1_score, recall_score, precision_score, confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt
import time
from sklearn.model_selection import train_test_split
import pandas as pd
```

```python
modelos = {
    'SVMLinear':SVC(kernel="linear"),
    'Bayes': GaussianNB(),
    'KNN':KNeighborsClassifier(n_neighbors=3),
    'CategorialBayes':CategoricalNB(),
    'DecisionTree':DecisionTreeClassifier(random_state=42),
    'MLP1CamadaOculta':MLPClassifier(hidden_layer_sizes=(12,16)),
    'MLP2CamadasOcultas':MLPClassifier(hidden_layer_sizes=(20,100)),
    }
```

```python


for nome_modelo, modelo in modelos.items():

    X_train_sub,X_test_sub,y_train_sub,y_test_sub = train_test_split(X_train, y_train, train_size=0.7, random_state=1)
    
    modelo.fit(X_train_sub,y_train_sub)
    with open(f'{nome_modelo}.pickle', 'wb') as f:
        pickle.dump(modelo,f)
    start = time.time()
    
    y_pred = modelo.predict(X_test_sub)
    print(time.time()-start)

    print(f"Modelo: {nome_modelo}")
    print(f"Acurácia: {accuracy_score(y_test_sub,y_pred):.3f}")
    print(f"Precisão: {precision_score(y_test_sub,y_pred):.3f}")
    print(f"Recall: {recall_score(y_test_sub,y_pred):.3f}")
    print(f"F1-Score: {f1_score(y_test_sub,y_pred):.3f}")
    disp = ConfusionMatrixDisplay(confusion_matrix=confusion_matrix(y_test_sub, y_pred, labels=clf.classes_),display_labels=clf.classes_)
    disp.plot()
    plt.show()
    
    print('----')
```

O melhor modelo foi MLP de 1 camada oculta com uma acurácia de 68.9% 3 uma precisão de 68.7%

```python
from sklearn.model_selection import cross_validate
```

```python
for nome_modelo, modelo in modelos.items():
    
    scores = cross_validate(modelo, X, y, scoring=['accuracy','f1'],cv=5)
    print(scores['test_accuracy'])
    
    print(f"Modelo: {nome_modelo}")
    print(f"Acurácia: {scores['test_accuracy'].mean():.3f} +- {scores['test_accuracy'].std():.2f}")
    print(f"Tempo de treino: {scores['fit_time'].mean():.4f} +- {scores['fit_time'].std():.4f}")
    print(f"Tempo de teste: {scores['score_time'].mean():.4f} +- {scores['score_time'].std():.4f}")
    print(f"F1-Score: {scores['test_f1'].mean():.2f} +- {scores['test_f1'].std():.2f}")
    
    print('----')

```
