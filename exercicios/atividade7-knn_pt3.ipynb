# 2
Utilize a técnica k-fold cross validation. Compare os resultados com o melhor resultado da Atividade 6.

```python
from sklearn.model_selection import cross_validate
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import recall_score,precision_score, f1_score, accuracy_score, classification_report
```

```python
scores = cross_validate(clf, X, y, scoring=['accuracy','f1','recall'],cv=10)
scores
```

```python
print(f"Resultado usando k-fold: {scores['test_accuracy'].mean()} +- {scores['test_accuracy'].std()}")
print(f"Tempo de treinamento usando k-fold: {scores['fit_time'].mean()} +- {scores['fit_time'].std()}")
print(f"Tempo de teste usando k-fold: {scores['score_time'].mean()} +- {scores['score_time'].std()}")
```

No modelo Hold out no exercicio 6 a acurácia foi 95%, já no K-fold foi de 98%, o que torna o K-fold mais preciso.

```python
from sklearn.pipeline import Pipeline
```

```python
X_train,X_test,y_train,y_test = train_test_split(X, y, train_size=0.9, random_state=1)
```

```python
X_train
```

```python
pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('clf', KNeighborsClassifier(n_neighbors = 7))
    ])
```

```python
X_train
```

```python
pipeline.fit(X_train, y_train)
```

```python
pipeline.score(X_test, y_test)
```

```python
y_pred = pipeline.predict(X_test)
accuracy_score(y_test,y_pred)
```
