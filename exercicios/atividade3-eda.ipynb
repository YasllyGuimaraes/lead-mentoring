```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
students_df = pd.read_csv('Student Mental health.csv')
students_df.head()
```

```python
dados_df = students_df.copy()
dados_df.head()
```

```python
# Para saber o título das colunas:
list(dados_df.columns)
```

```python

print(f"O conjunto de dados possui {len(dados_df)} amostras e {len(dados_df.columns)} colunas")
```

```python
dados_df.isna().sum()
```

Há apenas um dado faltante em na coluna Age.

```python
#Realizando limpeza de dados nulos : substituindo por No_data.
dados_df['Age'].fillna('No_data',inplace=True)
dados_df.head(44)
```

```python
#Agora não temos mais nenhum valor faltante
dados_df.dropna(inplace=True)
```

```python
len(dados_df)
```

Foi retirada a linha 43 que possuia o valor que não possuia dados.Após essa retirada o conjunto ficou com 100 amostras.

```python
dados_df.info()
```

```python
sb.countplot(x='Choose your gender',data=dados_df)
```

Com o gráfico acima podemos concluir que a maior parte dos estudantes que preencheram o formulário foram mulheres.

```python
dados_df['Choose your gender'].value_counts()
```

```python
sb.displot(data=dados_df, x="Choose your gender", hue="Do you have Depression?", col="Do you have Depression?")
```

De acordo com esse gráfico acima que correlaciona o indice de depressão dos alunos com o gênero é possível concluir que a maioria dos alunos, tanto homens quanto mulheres, não se consideram com depressão. Em contrapartida, ainda existem alunos e em sua maioria mulheres que se consideram com depressão.

```python


sb.histplot(data=dados_df,x='Your current year of Study')
```

A maioria das pessoas que reponderam a pesquisa estão no primeiro ano.

```python
#Colocando a coluna Timestamp em formato de data:
dados_df['Timestamp'] = pd.to_datetime(dados_df['Timestamp'])
dados_df.head()
```

```python
# Separando as horas da data em novas colunas:
dados_df['Hour added'] = dados_df['Timestamp'].dt.time
dados_df['Data added'] = dados_df['Timestamp'].dt.date
dados_df.head()
```

```python
ax = sb.histplot(data=dados_df,y='What is your course?')
ax.figure.set_size_inches(12, 12)
```

A maioria dos estudantes estão cursando BCS e Engenharia.

```python
ax = sb.stripplot(x='Do you have Depression?', y='What is your course?', data=dados_df, hue='Do you have Depression?', split=True)
ax.figure.set_size_inches(5, 12)
```

O índice de ataques de pânico se concentram mais nos cursos mais cursados: Engenharia, BIT e BCS.

```python
ax = sb.stripplot(x='Do you have Depression?', y='What is your CGPA?', data=dados_df, hue='Do you have Depression?', split=True)
ax.figure.set_size_inches(5, 12)
```

A maioria dos alunos que apresentam depressão são os alunos com as maiores médias.
